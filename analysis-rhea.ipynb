#### remove `$` and `,` -> transform the data into `floats` (doubles)

```python
degrees_that_pay['Starting Median Salary'] = degrees_that_pay['Starting Median Salary'].str.replace(',', '')
degrees_that_pay['Starting Median Salary'] = degrees_that_pay['Starting Median Salary'].str.replace('.', '')
degrees_that_pay['Starting Median Salary'] = degrees_that_pay['Starting Median Salary'].str.replace('$', '')
degrees_that_pay['Starting Median Salary'] = degrees_that_pay['Starting Median Salary'].astype(float)

degrees_that_pay['Mid-Career Median Salary'] = degrees_that_pay['Mid-Career Median Salary'].str.replace(',', '')
degrees_that_pay['Mid-Career Median Salary'] = degrees_that_pay['Mid-Career Median Salary'].str.replace('.', '')
degrees_that_pay['Mid-Career Median Salary'] = degrees_that_pay['Mid-Career Median Salary'].str.replace('$', '')
degrees_that_pay['Mid-Career Median Salary'] = degrees_that_pay['Mid-Career Median Salary'].astype(float)

degrees_that_pay['Mid-Career 10th Percentile Salary'] = degrees_that_pay['Mid-Career 10th Percentile Salary'].str.replace(',', '')
degrees_that_pay['Mid-Career 10th Percentile Salary'] = degrees_that_pay['Mid-Career 10th Percentile Salary'].str.replace('.', '')
degrees_that_pay['Mid-Career 10th Percentile Salary'] = degrees_that_pay['Mid-Career 10th Percentile Salary'].str.replace('$', '')
degrees_that_pay['Mid-Career 10th Percentile Salary'] = degrees_that_pay['Mid-Career 10th Percentile Salary'].astype(float)

degrees_that_pay['Mid-Career 25th Percentile Salary'] = degrees_that_pay['Mid-Career 25th Percentile Salary'].str.replace(',', '')
degrees_that_pay['Mid-Career 25th Percentile Salary'] = degrees_that_pay['Mid-Career 25th Percentile Salary'].str.replace('.', '')
degrees_that_pay['Mid-Career 25th Percentile Salary'] = degrees_that_pay['Mid-Career 25th Percentile Salary'].str.replace('$', '')
degrees_that_pay['Mid-Career 25th Percentile Salary'] = degrees_that_pay['Mid-Career 25th Percentile Salary'].astype(float)

degrees_that_pay['Mid-Career 75th Percentile Salary'] = degrees_that_pay['Mid-Career 75th Percentile Salary'].str.replace(',', '')
degrees_that_pay['Mid-Career 75th Percentile Salary'] = degrees_that_pay['Mid-Career 75th Percentile Salary'].str.replace('.', '')
degrees_that_pay['Mid-Career 75th Percentile Salary'] = degrees_that_pay['Mid-Career 75th Percentile Salary'].str.replace('$', '')
degrees_that_pay['Mid-Career 75th Percentile Salary'] = degrees_that_pay['Mid-Career 75th Percentile Salary'].astype(float)

degrees_that_pay['Mid-Career 90th Percentile Salary'] = degrees_that_pay['Mid-Career 90th Percentile Salary'].str.replace(',', '')
degrees_that_pay['Mid-Career 90th Percentile Salary'] = degrees_that_pay['Mid-Career 90th Percentile Salary'].str.replace('.', '')
degrees_that_pay['Mid-Career 90th Percentile Salary'] = degrees_that_pay['Mid-Career 90th Percentile Salary'].str.replace('$', '')
degrees_that_pay['Mid-Career 90th Percentile Salary'] = degrees_that_pay['Mid-Career 90th Percentile Salary'].astype(float)
```

```python
# interactive bar chart
salary_starting_median = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                                y=degrees_that_pay['Starting Median Salary'],
                                name='Starting Median Salary')

salary_mid_median = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                           y=degrees_that_pay['Mid-Career Median Salary'],
                           name='Mid-Career Median Salary',
                           visible=False)

salary_percentage_change = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                                  y=degrees_that_pay['Percent change from Starting to Mid-Career Salary'],
                                  name='Percent change from Starting to Mid-Career Salary',
                                  visible=False)

salary_mid_10th = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                           y=degrees_that_pay['Mid-Career 10th Percentile Salary'],
                           name='Mid-Career 10th Percentile Salary',
                           visible=False)

salary_mid_25th = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                           y=degrees_that_pay['Mid-Career 25th Percentile Salary'],
                           name='Mid-Career 25th Percentile Salary',
                           visible=False)

salary_mid_75th = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                           y=degrees_that_pay['Mid-Career 75th Percentile Salary'],
                           name='Mid-Career 75th Percentile Salary',
                           visible=False)

salary_mid_90th = go.Bar(x=degrees_that_pay['Undergraduate Major'], 
                           y=degrees_that_pay['Mid-Career 90th Percentile Salary'],
                           name='Mid-Career 90th Percentile Salary',
                           visible=False)

data = [salary_starting_median, 
        salary_mid_median,
        salary_percentage_change,
        salary_mid_10th,
        salary_mid_25th,
        salary_mid_75th,
        salary_mid_90th]

updatemenus = list([
    dict(active=-1,
         buttons=list([   
            dict(label = 'Starting Median Salary',
                 method = 'update',
                 args = [{'visible': [True, False, False, False, False, False, False]},
                         {'title': 'Starting Median Salary'}]),
            dict(label = 'Mid-Career Median Salary',
                 method = 'update',
                 args = [{'visible': [False, True, False, False, False, False, False]},
                         {'title': 'Mid-Career Median Salary'}]),
            dict(label = 'Percent change from Starting to Mid-Career Salary',
                 method = 'update',
                 args = [{'visible': [False, False, True, False, False, False, False]},
                         {'title': 'Percent change from Starting to Mid-Career Salary'}]),
            dict(label = 'Mid-Career 10th Percentile Salary',
                 method = 'update',
                 args = [{'visible': [False, False, False, True, False, False, False]},
                         {'title': 'Mid-Career 10th Percentile Salary'}]),
            dict(label = 'Mid-Career 25th Percentile Salary',
                 method = 'update',
                 args = [{'visible': [False, False, False, False, True, False, False]},
                         {'title': 'Mid-Career 25th Percentile Salary'}]),
            dict(label = 'Mid-Career 75th Percentile Salary',
                 method = 'update',
                 args = [{'visible': [False, False, False, False, False, True, False]},
                         {'title': 'Mid-Career 75th Percentile Salary'}]),
            dict(label = 'Mid-Career 90th Percentile Salary',
                 method = 'update',
                 args = [{'visible': [False, False, False, False, False, False, True]},
                         {'title': 'Mid-Career 90th Percentile Salary'}])
        ]),
    )
])

layout = dict(title='Undergraduate Major', showlegend=False,
              updatemenus=updatemenus)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='update_dropdown')
```
